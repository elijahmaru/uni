# Практическая работа №1

```python
users = ["  aNNA  ", "boris", "  ", "cLAra", "   daniEL ", "eva"]

clean_users = []

for u in users:
    if u == "  ":
        continue

    temp = ""
    for char in u:
        if char != " ":
            temp = temp + char

    if len(temp) > 0:
        first = temp[0].upper()
        rest = temp[1:].lower()
        fixed_name = first + rest
        clean_users.append(fixed_name)

print(f"Результат: {clean_users}")
```
