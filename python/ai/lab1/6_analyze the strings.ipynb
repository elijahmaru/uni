# Практическая работа №1

```python
raw_data = "Хлеб,5,40;Молоко,2,80;Сыр,1,350;Вода,10,25"

total_revenue = 0

items = raw_data.split(';')

for item in items:
    parts = item.split(',')

    qty_str = parts[1]
    price_str = parts[2]

    qty = int(qty_str)
    price = int(price_str)

    current_sum = qty * price
    total_revenue = total_revenue + current_sum

print(f"Общая выручка: {total_revenue} руб")
```
