```python
p = "MyPass123!"
a = False
b = False
c = False
d = False
if len(p) >= 8:
    a = True
for ch in p:
    if ch.isupper():
        b = True
    if ch.islower():
        c = True
    if ch.isdigit():
        d = True
if a and b and c and d:
    print("Valid password")
else:
    print("Invalid password")
```
