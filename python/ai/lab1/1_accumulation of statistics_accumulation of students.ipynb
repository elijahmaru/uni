### Практическая работа №1 ###

```python
grades = [85, 92, 78, 96, 88, 75, 90, 82, 95, 87]
cnt = 0
sm = 0
hi = grades[0]
lo = grades[0]
for g in grades:
    cnt = cnt + 1
    sm = sm + g
    if g > hi:
        hi = g
    if g < lo:
        lo = g
avg = sm / cnt
print(f"Count: {cnt}")
print(f"Average: {avg:.2f}")
print(f"Highest: {hi}")
print(f"Lowest: {lo}")
```
