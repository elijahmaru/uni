```python
distance_km = 450
avg_speed_kmh = 90
fuel_consumption_per_100 = 8.5 # литров
gas_price = 55.0 # рублей за литр

print("--- Расчет поездки ---")

time_hours = distance_km / avg_speed_kmh

required_fuel = (distance_km / 100) * fuel_consumption_per_100

total_cost = required_fuel * gas_price

print(f"Дистанция: {distance_km} км")
print(f"Время в пути: {time_hours} ч")
print(f"Потрачено бензина: {required_fuel} л")
print(f"Стоимость поездки: {total_cost} руб")

# Задача: разбить этот код на функции, чтобы он стал переиспользуемым
```
