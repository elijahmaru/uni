### Практическая работа №1 ###

```python
s = "programming is fun"
d = {}
for c in s:
    if c == ' ':
        continue
    if c in d:
        d[c] = d[c] + 1
    else:
        d[c] = 1
m = 0
k = ''
for x in d:
    if d[x] > m:
        m = d[x]
        k = x
print(f"Most frequent: {k} appears {m} times")
```
