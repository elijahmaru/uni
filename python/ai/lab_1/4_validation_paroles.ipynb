```python
products = ["apple", "banana", "orange", "milk"]
stocks = [10, 5, 20, 8]

def get_stock(product_name):
    found_index = -1
    for i in range(len(products)):
        if products[i] == product_name:
            found_index = i
            break

    if found_index != -1:
        return stocks[found_index]
    else:
        return 0

item = "banana"
qty = get_stock(item)
print(f"На складе {item}: {qty} шт.")
```
