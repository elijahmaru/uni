Рассмотрим основные структуры данных в языке программирования Python: списки (lists), кортежи (tuples), словари (dictionaries), множества (sets), строки (strings), а также понятия изменяемых и неизменяемых типов данных.

## 1. Списки (Lists)

Списки - это упорядоченные изменяемые коллекции объектов произвольных типов. Они позволяют хранить несколько элементов в одном объекте и поддерживают индексацию и срезы.

```python
# Создание списка
my_list = [1, 2, 3, 4, 5]
print("Созданный список:", my_list)
```

```python
# Операции со списками
print("Исходный список:", my_list)

# Добавление элементов
my_list.append(6)
print("После добавления элемента 6:", my_list)

# Вставка элемента по индексу
my_list.insert(2, 10)
print("После вставки элемента 10 на позицию 2:", my_list)

# Удаление элемента по значению
my_list.remove(10)
print("После удаления элемента 10:", my_list)

# Удаление элемента по индексу
removed_element = my_list.pop(0)
print(f"Удалённый элемент: {removed_element}")
print("После удаления первого элемента:", my_list)
```

```python
# Индексация и срезы
print("Список:", my_list)
print("Первый элемент (индекс 0):", my_list[0])
print("Последний элемент (индекс -1):", my_list[-1])
print("Срез от 1 до 4 (не включая 4):", my_list[1:4])
print("Срез от начала до 3-го элемента:", my_list[:3])
print("Срез от 2-го элемента до конца:", my_list[2:])
print("Срез с шагом 2:", my_list[::2])
print("Обратный порядок:", my_list[::-1])
```

```python
# Методы списков
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print("Исходный список:", numbers)

# Сортировка
numbers.sort()
print("Отсортированный список:", numbers)

# Разворот списка
numbers.reverse()
print("Развёрнутый список:", numbers)

# Количество вхождений элемента
count = numbers.count(1)
print(f"Количество вхождений числа 1: {count}")

# Поиск индекса элемента
index = numbers.index(5)
print(f"Индекс первого вхождения числа 5: {index}")

# Расширение списка
numbers.extend([7, 8, 9])
print("После расширения:", numbers)
```

## 2. Кортежи (Tuples)

Кортежи - это упорядоченные неизменяемые коллекции объектов произвольных типов. Они работают как списки, но не могут быть изменены после создания.

```python
# Создание кортежа
my_tuple = (1, 2, 3, 4, 5)
print("Созданный кортеж:", my_tuple)

# Создание кортежа без скобок
another_tuple = 1, 2, 3
print("Ещё один кортеж:", another_tuple)

# Создание кортежа с одним элементом
single_tuple = (1,)
print("Кортеж с одним элементом:", single_tuple)
```

```python
# Операции с кортежами
print("Кортеж:", my_tuple)

# Индексация и срезы (работают так же как у списков)
print("Первый элемент:", my_tuple[0])
print("Последние два элемента:", my_tuple[-2:])

# Методы кортежей
print("Количество вхождений 3:", my_tuple.count(3))
print("Индекс элемента 4:", my_tuple.index(4))

# Конкатенация кортежей
combined_tuple = my_tuple + (6, 7)
print("Комбинированный кортеж:", combined_tuple)

# Умножение кортежа
multiplied_tuple = my_tuple * 2
print("Удвоенный кортеж:", multiplied_tuple)
```

## 3. Словари (Dictionaries)

Словари - это неупорядоченные изменяемые коллекции пар ключ-значение. Ключи должны быть уникальными и неизменяемыми.

```python
# Создание словаря
my_dict = {'name': 'Alice', 'age': 25, 'city': 'Moscow'}
print("Созданный словарь:", my_dict)

# Создание словаря с помощью dict()
another_dict = dict(a=1, b=2, c=3)
print("Другой словарь:", another_dict)
```

```python
# Операции со словарями
print("Исходный словарь:", my_dict)

# Доступ к значениям по ключу
print("Имя:", my_dict['name'])

# Добавление/изменение элементов
my_dict['country'] = 'Russia'
print("После добавления страны:", my_dict)

# Удаление элемента
del my_dict['age']
print("После удаления возраста:", my_dict)

# Получение значения с умолчанием
phone = my_dict.get('phone', 'Не указан')
print(f"Телефон: {phone}")
```

```python
# Методы словарей
print("Словарь:", my_dict)

# Получение всех ключей
keys = list(my_dict.keys())
print("Ключи:", keys)

# Получение всех значений
values = list(my_dict.values())
print("Значения:", values)

# Получение всех пар ключ-значение
items = list(my_dict.items())
print("Пары ключ-значение:", items)

# Обновление словаря
my_dict.update({'age': 30, 'job': 'Engineer'})
print("После обновления:", my_dict)
```

## 4. Множества (Sets)

Множества - это неупорядоченные изменяемые коллекции уникальных элементов. Они полезны для математических операций над множествами.

```python
# Создание множества
my_set = {1, 2, 3, 4, 5}
print("Созданное множество:", my_set)

# Создание множества с помощью set()
another_set = set([1, 2, 2, 3, 3, 4])
print("Множество из списка (без дубликатов):", another_set)
```

```python
# Операции с множествами
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(f"Множество 1: {set1}")
print(f"Множество 2: {set2}")

# Объединение
union_set = set1 | set2
print(f"Объединение: {union_set}")

# Пересечение
intersection_set = set1 & set2
print(f"Пересечение: {intersection_set}")

# Разность
difference_set = set1 - set2
print(f"Разность (set1 - set2): {difference_set}")

# Симметрическая разность
symmetric_diff = set1 ^ set2
print(f"Симметрическая разность: {symmetric_diff}")
```

```python
# Методы множеств
my_set = {1, 2, 3}
print("Исходное множество:", my_set)

# Добавление элемента
my_set.add(4)
print("После добавления 4:", my_set)

# Удаление элемента (ошибка, если элемента нет)
my_set.remove(2)
print("После удаления 2:", my_set)

# Удаление элемента (без ошибки, если элемента нет)
my_set.discard(10)
print("После discard(10):", my_set)

# Проверка принадлежности
print(f"Содержит ли 3? {3 in my_set}")

# Подмножество и надмножество
set_a = {1, 2}
set_b = {1, 2, 3, 4}
print(f"{set_a} подмножество {set_b}? {set_a.issubset(set_b)}")
print(f"{set_b} надмножество {set_a}? {set_b.issuperset(set_a)}")
```

## 5. Строки (Strings)

Строки - это неизменяемые последовательности символов. Они поддерживают индексацию и срезы, как списки.

```python
# Создание строк
my_string = "Hello, World!"
print("Созданная строка:", my_string)

# Многострочные строки
multiline_string = """Это многострочная
строка, которая может
содержать несколько строк."""
print("Многострочная строка:", multiline_string)
```

```python
# Операции со строками
text = "Python Programming"
print("Исходная строка:", text)

# Индексация и срезы
print("Первый символ:", text[0])
print("Последний символ:", text[-1])
print("Срез от 0 до 6:", text[0:6])
print("Обратный порядок:", text[::-1])

# Конкатенация
greeting = "Hello, " + text[:6]
print("Приветствие:", greeting)

# Умножение строки
repeated = "Hi! " * 3
print("Повторенная строка:", repeated)
```

```python
# Методы строк
text = "  Hello, World!  "
print("Исходная строка: '" + text + "'")

# Удаление пробелов
stripped = text.strip()
print("После strip(): '" + stripped + "'")

# Преобразование регистра
upper_text = text.upper().strip()
lower_text = text.lower().strip()
print("Верхний регистр:", upper_text)
print("Нижний регистр:", lower_text)

# Замена подстроки
replaced = text.replace('World', 'Python').strip()
print("После замены:", replaced)

# Разбиение строки
words = "apple,banana,orange".split(',')
print("Разбитие по запятой:", words)

# Соединение строк
joined = '-'.join(words)
print("Соединение через дефис:", joined)

# Поиск подстроки
position = text.find('World')
print(f"Позиция слова 'World': {position}")
```

## 6. Изменяемые и неизменяемые типы данных

### Изменяемые (mutable):
- Списки (list)
- Словари (dict)
- Множества (set)
- bytearray

### Неизменяемые (immutable):
- Числа (int, float, complex)
- Строки (str)
- Кортежи (tuple)
- frozenset
- bytes

```python
# Пример изменяемого типа (список)
original_list = [1, 2, 3]
reference_list = original_list

print(f"Оригинальный список: {original_list}")
print(f"Ссылка на список: {reference_list}")

# Изменяем оригинальный список
original_list.append(4)
print(f"После изменения оригинала: {original_list}")
print(f"Ссылка также изменилась: {reference_list}")
```

```python
# Пример неизменяемого типа (строка)
original_string = "Hello"
reference_string = original_string

print(f"Оригинальная строка: {original_string}")
print(f"Ссылка на строку: {reference_string}")

# Создаем новую строку (операция создает новый объект)
new_string = original_string + " World"
print(f"Новая строка: {new_string}")
print(f"Оригинальная строка не изменилась: {original_string}")
print(f"Ссылка не изменилась: {reference_string}")
```

## 7. Упаковка в кортежи (Tuple Packing и Unpacking)

Python позволяет упаковывать несколько значений в кортеж и распаковывать их обратно.

```python
# Упаковка в кортеж (tuple packing)
packed_tuple = 1, 2, 3, 4
print("Упакованный кортеж:", packed_tuple)

# Распаковка кортежа (tuple unpacking)
a, b, c, d = packed_tuple
print(f"Распаковка: a={a}, b={b}, c={c}, d={d}")

# Распаковка с использованием *
first, *middle, last = packed_tuple
print(f"Распаковка с *: first={first}, middle={middle}, last={last}")

# Использование распаковки при обмене значений
x, y = 10, 20
print(f"До обмена: x={x}, y={y}")
x, y = y, x  # Обмен значениями с помощью распаковки
print(f"После обмена: x={x}, y={y}")
```

## 8. Преобразования между структурами данных

Python предоставляет встроенные функции для преобразования между различными структурами данных.

```python
# Преобразование списка в кортеж
list_example = [1, 2, 3, 4, 5]
tuple_from_list = tuple(list_example)
print(f"Список: {list_example}")
print(f"Кортеж из списка: {tuple_from_list}")

# Преобразование кортежа в список
tuple_example = (6, 7, 8, 9, 10)
list_from_tuple = list(tuple_example)
print(f"Кортеж: {tuple_example}")
print(f"Список из кортежа: {list_from_tuple}")

# Преобразование строки в список
string_example = "hello"
list_from_string = list(string_example)
print(f"Строка: {string_example}")
print(f"Список символов: {list_from_string}")

# Преобразование списка в строку
char_list = ['h', 'e', 'l', 'l', 'o']
string_from_list = ''.join(char_list)
print(f"Список символов: {char_list}")
print(f"Строка: {string_from_list}")
```

```python
# Преобразование списка в множество
list_with_duplicates = [1, 2, 2, 3, 3, 4, 5, 5]
set_from_list = set(list_with_duplicates)
print(f"Список с дубликатами: {list_with_duplicates}")
print(f"Множество (без дубликатов): {set_from_list}")

# Преобразование множества в список
set_example = {1, 2, 3, 4, 5}
list_from_set = list(set_example)
print(f"Множество: {set_example}")
print(f"Список из множества: {list_from_set}")

# Преобразование списка пар в словарь
pairs_list = [('name', 'Alice'), ('age', 25), ('city', 'Moscow')]
dict_from_pairs = dict(pairs_list)
print(f"Список пар: {pairs_list}")
print(f"Словарь: {dict_from_pairs}")

# Преобразование словаря в список ключей
keys_list = list(dict_from_pairs.keys())
print(f"Список ключей: {keys_list}")

# Преобразование словаря в список значений
values_list = list(dict_from_pairs.values())
print(f"Список значений: {values_list}")

# Преобразование словаря в список пар ключ-значение
items_list = list(dict_from_pairs.items())
print(f"Список пар ключ-значение: {items_list}")
```

## 9. Практические примеры использования структур данных

Рассмотрим несколько практических примеров использования различных структур данных.

```python
# Пример: подсчёт частоты слов в тексте с использованием словаря
text = "python is great python is powerful python is easy to learn"
words = text.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print("Частота слов:", word_count)

# Пример: поиск уникальных элементов с использованием множества
numbers_with_duplicates = [1, 2, 3, 2, 4, 5, 3, 6, 1, 7, 8, 5]
unique_numbers = list(set(numbers_with_duplicates))
print("Оригинальный список:", numbers_with_duplicates)
print("Уникальные числа:", sorted(unique_numbers))

# Пример: использование кортежей как координат для представления точек
points = [(0, 0), (1, 2), (3, 4), (2, 1)]
print("Список точек (координаты):", points)

# Вычисление расстояния от начала координат для каждой точки
import math
distances = []
for x, y in points:
    distance = math.sqrt(x**2 + y**2)
    distances.append((x, y, round(distance, 2)))
    
print("Точки с расстоянием от начала координат:", distances)
```

## 9.1 Применение метода lower() для приведения слов к нижнему регистру

Использование метода .lower() позволяет избежать проблем с наличием слов с заглавной буквы, приводя все слова к нижнему регистру перед обработкой.

```python
# Пример: подсчёт частоты слов в тексте с использованием .lower() для нормализации регистра
text = "Python is Great Python is Powerful python is Easy to Learn"
words = text.split()
print("Слова в исходном виде:", words)

# Применение .lower() для приведения всех слов к нижнему регистру
normalized_words = [word.lower() for word in words]
print("Слова после применения .lower():", normalized_words)

# Подсчет частоты слов с нормализованными данными
word_count = {}
for word in normalized_words:
    word_count[word] = word_count.get(word, 0) + 1

print("Частота слов после нормализации:", word_count)

# Сравнение: без .lower() те же слова с разным регистром будут считаться как разные
words_original = text.split()
word_count_original = {}
for word in words_original:
    word_count_original[word] = word_count_original.get(word, 0) + 1

print("Частота слов без нормализации (разный регистр влияет на результат):", word_count_original)
```
